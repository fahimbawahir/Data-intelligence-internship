```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
df = pd.read_csv("renewable-share-energy.csv")
focus = df[df["Entity"].isin(["United States", "India", "Germany", "Brazil"])]
 
plt.figure(figsize=(10, 6))
sns.lineplot(data=focus, x="Year",
             y="Renewables",
             hue="Entity", linewidth=2)
plt.title("Renewable share is rising everywhere — at very different speeds")
plt.xlabel("Year")
plt.ylabel("Renewables (% of primary energy)")
plt.legend(title=None)
plt.tight_layout()
plt.show()
```

```python
df_2024 = df[df["Year"] == 2024]

top10 = df_2024.sort_values(
    by="Renewables",
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

plt.bar(top10["Entity"], top10["Renewables"])

plt.title("Top 10 Countries by Renewable Energy Share (2024)")
plt.xlabel("Country")
plt.ylabel("Renewable Share (%)")

plt.xticks(rotation=45)


plt.show()
```

```python
df_2024 = df[df["Year"] == 2024]

plt.figure(figsize=(10,6))

plt.hist(df_2024["Renewables"], bins=20)

plt.title("Distribution of Renewable Energy Share Across Countries (2024)")
plt.xlabel("Renewable Energy Share (%)")
plt.ylabel("Number of Countries")

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("merged_renewable_gdp.csv")

corr_data = df[[
    "renewables_pct",
    "GDP_per_capita"
]]

corr_data = corr_data.dropna()

corr_matrix = corr_data.corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.tight_layout()
plt.show()
```
