```python
principal = float(input("enter principal amount: "))
rate =  float(input("enter compound rate as a decimal: "))
years = int(input("enter number of years: "))
future_val = 0
def compound_growth(principal, rate, years):
    '''Future value of a principal at a constant compounding rate.'''
    future_val = principal * (1+rate)**years
    return future_val
    
compound_growth(principal, rate, years)  
```

```python
def classify_growth(rate):

    results = []

    for n in rate:

        if n < 0:
            results.append("recession")
        elif n > 7:
            results.append("overheated")
        elif n > 0 and n <= 2:
            results.append("slow")
        elif n > 2 and n <= 7:
            results.append("healthy")
    return results


print(classify_growth([2, 3, 9, 11]))
```
