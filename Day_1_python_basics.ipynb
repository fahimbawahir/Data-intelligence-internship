```python
#9
gdp_list = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
total = 0
counter = 0
for m in gdp_list:
    total += m
    counter += 1
mean_gdp = total/counter

for n in gdp_list:
    if n > mean_gdp:
        print(n)
```

```python
renewables = {
    "Norway": 71.6, 
    "Brazil": 45.0,
    "Germany": 19.7,
    "USA": 11.4,
    "India": 9.8,
    "Sweden": 50.2,
    "France": 13.1,
    "UK": 18.2,
}

total = 0
count = 0
for country, share in renewables.items():
    total += share
    count +=1

mean = total/count

for country, share in renewables.items():
    if share > mean:
        print(country)
```
